```python
from pyspark import SparkConf, SparkContext
from operator import add
sc = SparkContext(conf=SparkConf().setAppName("MyApp").setMaster("yarn"))
```

```python
import re

def parse_article(line):
    try:
        article_id, text = line.rstrip().split('\t', 1)
        text = re.sub("^\W+|\W+$", "", text, flags=re.UNICODE)
        words = re.split("\W*\s+\W*", text, flags=re.UNICODE)
        return words
    except ValueError as e:
        return []
    
def get_word_pairs(words):
    pairs = []
    for i in range(len(words)-1):
        pairs.append((words[i].lower()+'_'+words[i+1].lower(), 1))
    return pairs

wiki = sc.textFile("/data/wiki/en_articles_part/articles-part", 4).map(parse_article)
word_pairs = wiki.flatMap(get_word_pairs)
word_pairs = word_pairs.filter(lambda kv: kv[0].startswith('narodnaya'))
```

```python
result = word_pairs.foldByKey(0, add)
```

```python
result = result.sortByKey().collect()
```

```python
sc.stop()
```

```python
for k,v in result:
    print("{}\t{}".format(k,v))
```
